## First Law of Thermodynamics
$ΔE = Q - W,$

where ΔE is the change in the internal energy of a System, Q is the heat exchanged
with a reservoir, and W is the work done by the System. In any spontaneous process,
a System evolves to the lowest accessible State of energy. If this State is time
independent, then we say that the System is in equilibrium.
